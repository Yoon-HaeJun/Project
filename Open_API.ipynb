```python
import requests
import os

API_KEY = os.getenv("ECOS_API_KEY")

STAT_CODE = "731Y001"
ITEM_CODE = "0000002"

START_DATE = "20210101"
END_DATE = "20251231"

records = []

start = 1
step = 1000

while True:
    end = start + step -1
    
    url = f"https://ecos.bok.or.kr/api/StatisticSearch/{API_KEY}/json/kr/{start}/{end}/{STAT_CODE}/D/{START_DATE}/{END_DATE}"


    response = requests.get(url, timeout = 60)
    response.raise_for_status()
    data = response.json()

    stat = data.get("StatisticSearch", {})
        
    if "RESULT" in stat:
        print("ECOS 에러:", stat["RESULT"])
        break

    rows = stat.get("row", [])

    if not rows:
        break

    for r in rows:
        if r["ITEM_CODE1"] == ITEM_CODE:
            records.append({
                "date" : r["TIME"],
                "value" : float(r["DATA_VALUE"])
            })

    print(f"{start} ~ {end} 수집 완료 (누적 {len(records)})")
    
    start += step

print("전체 수집 완료:", len(records))
```
